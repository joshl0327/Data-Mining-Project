```python
import methods as m
```

```python
start = '2007-01-01'
end = '2018-01-01'

tech_names = ['AAPL', 'MSFT', 'GOOGL', 'FB', 'INTC']
pub_util_names = ['AMZN', 'WMT', 'HD', 'CMCSA', 'DIS']
health_names = ['JNJ', 'UNH', 'PFE', 'MRK', 'ABBV']
finance_names = ['JPM', 'BAC', 'WFC', 'C', 'AXP']
energy_names = ['XOM', 'CVX', 'GE', 'BP', 'COP']
cons_serv_names = ['VZ', 'T', 'NEE', 'DUK', 'TMUS']
cons_nd_names = ['KO', 'PEP', 'MO', 'PM', 'NKE']
capital_names = ['BA', 'UTX', 'HON', 'TMO', 'LMT']

spy = m.get_spy('yahoo',start,end)
```

```python
industries = [tech_names, pub_util_names, health_names, finance_names, energy_names, 
              cons_serv_names, cons_nd_names, capital_names]
for x in industries:
    stocker_stocks = m.get_stocker(x)
    stock_close_df = m.get_multiple_close(x, stocker_stocks, start, end)
    stock_spy = m.add_spy(spy, stock_close_df)
    stock_names_spy = x
    stock_names_spy.append('SPY')
    m.plot_multiple(stock_names_spy, stock_spy, plot_type='returns')
    
    stock_close_df_2012 = m.get_multiple_close(x, stocker_stocks, start, '2012-01-01')
    spy_vol = m.get_spy('yahoo',start,'2012-01-01')
    stock_close_df_2012 = m.add_spy(spy_vol, stock_close_df_2012)
    m.volatility(stock_close_df_2012)
    
    stock_close_df_cp = m.get_multiple_close(x, stocker_stocks, '2015-03-30', '2018-03-27')
    spy_vol = m.get_spy('yahoo','2015-03-30','2018-03-27')
    stock_close_df_cp = m.add_spy(spy_vol, stock_close_df_cp)
    m.volatility(stock_close_df_cp)
```
